# Variable Descriptions

- **`EmployeeID`**: A unique identifier for each employee.

- **`Department`**: The department where the employee works, such as Sales, Research & Development, or Human Resources.

- **`JobRole`**: The specific role or title of the employee, such as:

  - Human Resources

- **`MaritalStatus`**: The marital status of the employee, e.g., Married, Single, Divorced.

- **`NumCompaniesWorked`**: The number of companies the employee has worked for prior to the current company.

- **`Over18`**: Indicates whether the employee is over 18 years of age (often a binary field, usually always "Y").

# Attrition by Department Analysis

```python
# Calculate total employees in each department
total_by_department = df.groupby('Department').size()

# Calculate total attrition 'Yes' and 'No' in each department
attrition_yes_by_department = df[df['Attrition'] == 'Yes'].groupby('Department').size()
attrition_no_by_department = df[df['Attrition'] == 'No'].groupby('Department').size()

# Calculate attrition rates for 'Yes' and 'No' as percentage
attrition_yes_rate_by_department = (attrition_yes_by_department / total_by_department) * 100
attrition_no_rate_by_department = (attrition_no_by_department / total_by_department) * 100

# Combine into a DataFrame for plotting
attrition_rate_df = pd.DataFrame({
    'Attrition Yes (%)': attrition_yes_rate_by_department,
    'Attrition No (%)': attrition_no_rate_by_department
})

# Plot a grouped bar chart
ax = attrition_rate_df.plot(kind='bar', figsize=(12, 6), color=['salmon', 'lightblue'])
plt.title('Attrition Rates by Department (Yes vs No)')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Attrition Status')

# Annotate bars with percentage values
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fmt='%.1f%%', fontsize=10, padding=3)

plt.show()
```

📌 **From above Bar Plot we can say that employees working in human resource are 30.2% more likely to leave which is almost double compared to other departments.**

# Distance from Home Distribution responsible for Employees Attrition

```python
# Set the figure size
plt.figure(figsize=(10, 6))  

# Create a stacked histogram
sns.histplot(data=df, x='DistanceFromHome', hue='Attrition', multiple='stack', bins=30, palette='Set2')

# Set the title and labels
plt.title('Distance from Home Distribution for Employees with Attrition Status')  # Set the title
plt.xlabel('Distance from Home')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label

# Set custom legend labels
plt.legend(title='Attrition Status', labels=['Still Employed', 'Left Company'])  # Set the legend title and labels

# Display the plot
plt.show()  
```

```python
# Split the DataFrame into two based on Attrition
df_yes = df[df['Attrition'] == 'Yes'].copy()  # Make a copy to avoid SettingWithCopyWarning
df_no = df[df['Attrition'] == 'No'].copy()    # Make a copy to avoid SettingWithCopyWarning

# Define the bins and labels
bins = [0, 5, 10, 15, 20, 25, 30]
labels = ['0-5', '5-10', '10-15', '15-20', '20-25', '25-30']

# Create a new column with binned distances for attrition = Yes
df_yes.loc[:, 'DistanceBin'] = pd.cut(df_yes['DistanceFromHome'], bins=bins, labels=labels, right=False)

# Count the number of employees in each bin for attrition = Yes
bin_counts_yes = df_yes['DistanceBin'].value_counts().sort_index()

# Calculate total attrition 'Yes' count
total_attrition_yes = len(df_yes)

# Calculate the percentage for each bin for attrition = Yes
percentage_yes = (bin_counts_yes / total_attrition_yes) * 100

# Create a result DataFrame for attrition = Yes
result_yes = percentage_yes.reset_index()
result_yes.columns = ['Distance Bin', 'Percentage']
result_yes['Count'] = bin_counts_yes.values

# Display results for attrition = Yes
print("Attrition = Yes:")
print(result_yes)

# Create a new column with binned distances for attrition = No
df_no.loc[:, 'DistanceBin'] = pd.cut(df_no['DistanceFromHome'], bins=bins, labels=labels, right=False)

# Count the number of employees in each bin for attrition = No
bin_counts_no = df_no['DistanceBin'].value_counts().sort_index()

# Calculate total attrition 'No' count
total_attrition_no = len(df_no)

# Calculate the percentage for each bin for attrition = No
percentage_no = (bin_counts_no / total_attrition_no) * 100

# Create a result DataFrame for attrition = No
result_no = percentage_no.reset_index()
result_no.columns = ['Distance Bin', 'Percentage']
result_no['Count'] = bin_counts_no.values

# Display results for attrition = No
print("\nAttrition = No:")
print(result_no)
```

📌**From above plot we can conclude the employees at distance from company betwenn 0-5 are more likely to leave thae company as shown by the count of 273.**

```python
# Calculate mean and median for MonthlyIncome by JobRole
mean_salary = df.groupby('JobRole')['MonthlyIncome'].mean().reset_index()
median_salary = df.groupby('JobRole')['MonthlyIncome'].median().reset_index()

# Rename the columns for clarity
mean_salary.columns = ['JobRole', 'MeanMonthlyIncome']
median_salary.columns = ['JobRole', 'MedianMonthlyIncome']

# Merge the mean and median dataframes for a combined view
salary_stats = pd.merge(mean_salary, median_salary, on='JobRole')

# Display the results
print(salary_stats)
```

```python
# Create a function to identify outliers and count them
def count_outliers(group):
    Q1 = group['MonthlyIncome'].quantile(0.25)
    Q3 = group['MonthlyIncome'].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers
    outliers_count = group[(group['MonthlyIncome'] < lower_bound) | (group['MonthlyIncome'] > upper_bound)].shape[0]
    return pd.Series({'OutliersCount': outliers_count})

# Group by JobRole and apply the outlier count function
outliers_summary = df.groupby('JobRole').apply(count_outliers).reset_index()

# Display the outliers summary
print(outliers_summary)
```

📌**From above plots and Mean and Median Values for monthly salary we can say that the job role of Manufacturing Director is the most highest paying and stable job with least outliers while the monthly salary for job role of Research Scientist is more dependant on other factors.**

# Effect of Job Employment on Employee Attrition

```python
# Create a DataFrame with counts of JobInvolvement and Attrition
counts = df.groupby(['JobInvolvement', 'Attrition']).size().reset_index(name='Count')

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the bar chart
bar_plot = sns.barplot(data=counts, x='JobInvolvement', y='Count', hue='Attrition', palette='Set2')

# Add labels and title
plt.title('Job Involvement vs. Attrition')
plt.xlabel('Job Involvement')
plt.ylabel('Number of Employees')

# Calculate total counts for each Job Involvement level
total_counts = counts.groupby('JobInvolvement')['Count'].sum().reset_index(name='TotalCount')

# Loop through each job involvement level to add percentage labels
for i in range(len(total_counts)):
    job_involvement = total_counts['JobInvolvement'].iloc[i]
    total = total_counts['TotalCount'].iloc[i]
    
    # Get the counts for attrition Yes and No
    attrition_yes_count = counts[(counts['JobInvolvement'] == job_involvement) & (counts['Attrition'] == 'Yes')]['Count'].sum()
    attrition_no_count = counts[(counts['JobInvolvement'] == job_involvement) & (counts['Attrition'] == 'No')]['Count'].sum()
    
    # Calculate percentages
    percent_attrition_yes = (attrition_yes_count / total) * 100 if total > 0 else 0
    percent_attrition_no = (attrition_no_count / total) * 100 if total > 0 else 0
    
    # Get the position of the bars to place the labels correctly
    y_offset_yes = attrition_yes_count + 20.0  # Slightly above the 'Yes' bar
    y_offset_no = attrition_no_count + 20.0   # Slightly above the 'No' bar
    
    # Place text labels for Yes and No
    plt.text(i - 0.2, y_offset_no, f'{percent_attrition_no:.1f}%', color='black', ha='center')
    plt.text(i + 0.2, y_offset_yes, f'{percent_attrition_yes:.1f}%', color='black', ha='center')

# Show the legend
plt.legend(title='Attrition')

plt.show()
```

📌 **From the above bar plot we can conclude the Employees with more job involvement of 1 and 4 are highly likely to leave (i.e., 21.7% and 18.1% respectively) the company compared to others and most employees have job involvement rating of 3.**

# Average Monthly Income by Employee Education Level to find out which Employees are more likely to leave the company

```python
# Calculate average monthly income by education level and attrition status
avg_salary_attrition = df.groupby(['Education', 'Attrition'])['MonthlyIncome'].mean().reset_index()

# Calculate count for each group
count_salary_attrition = df.groupby(['Education', 'Attrition']).size().reset_index(name='Count')

# Merge average and count dataframes
plot_data = avg_salary_attrition.merge(count_salary_attrition, on=['Education', 'Attrition'])

# Create a bar plot
plt.figure(figsize=(14, 8))
ax = sns.barplot(data=plot_data, x='Education', y='MonthlyIncome', hue='Attrition', palette='Set2')

# Add labels and title
plt.title('Average Monthly Income by Education Level and Attrition', fontsize=16)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Average Monthly Income', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Attrition', title_fontsize='12', fontsize='10')

# Function to add value labels
def add_value_labels(ax, spacing=5):
    for rect in ax.patches:
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Find the corresponding count
        education = plot_data['Education'][int(x_value - 0.25)]
        attrition = 'Yes' if x_value % 1 > 0.5 else 'No'
        count = plot_data[(plot_data['Education'] == education) & (plot_data['Attrition'] == attrition)]['Count'].values[0]

        # Add count and average income labels
        ax.text(x_value, y_value, f'${y_value:,.0f}\nn={count}', 
                ha='center', va='bottom', fontweight='bold', fontsize=9)

# Add value labels
add_value_labels(ax)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

📌 **From above grouped bar chart we can conclude that Employees with Lowest and Highest Education Level (i.e., 5) are more likely to switch jobs inspite of having high salaries (i.e., $63838  and  $74063) compared to their other counterpart.**

# Business Travel vs Depatment and Job Role Grouped Bar Plot

```python
# Create a figure with 2 subplots, one for each plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# BusinessTravel vs Department Grouped Bar Plot
grouped_data_dept = df.groupby(['Department', 'BusinessTravel']).size().unstack(fill_value=0)
departments = grouped_data_dept.index
business_travel = grouped_data_dept.columns

x = np.arange(len(departments))  # The label locations
width = 0.25  # The width of the bars

# Plotting each 'BusinessTravel' category as a separate bar for Department
for i, bt in enumerate(business_travel):
    ax1.bar(x + i * width, grouped_data_dept[bt], width, label=bt)

# Adding labels and title for the first plot
ax1.set_xlabel('Department')
ax1.set_ylabel('Count')
ax1.set_title('Business Travel vs Department Grouped Bar Plot')
ax1.set_xticks(x + width)
ax1.set_xticklabels(departments)
ax1.legend(title="Business Travel")

# BusinessTravel vs JobRole Grouped Bar Plot
grouped_data_job = df.groupby(['JobRole', 'BusinessTravel']).size().unstack(fill_value=0)
job_roles = grouped_data_job.index

# Insert line breaks to split the job roles into multiple lines
job_roles_split = [role.replace(' ', '\n') for role in job_roles]

x2 = np.arange(len(job_roles))  # The label locations

# Plotting each 'BusinessTravel' category as a separate bar for JobRole
for i, bt in enumerate(business_travel):
    ax2.bar(x2 + i * width, grouped_data_job[bt], width, label=bt)

# Adding labels and title for the second plot
ax2.set_xlabel('Job Role')
ax2.set_ylabel('Count')
ax2.set_title('Business Travel vs Job Role Grouped Bar Plot')
ax2.set_xticks(x2 + width)
ax2.set_xticklabels(job_roles_split)  # Using the split job role labels
ax2.legend(title="Business Travel")

# Adjust layout to prevent overlap and display the plots
plt.tight_layout()
plt.show()
```

📌 **From above plots we can observe that a major chunk of employees have to travel rarely irrespective of departments or roles followed by the count of those employees who travel frequently and then never travel respectively.***
