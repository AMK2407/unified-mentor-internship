# Introduction

A bird strike is strictly defined as a collision between a bird and an aircraft which is in
flight or on a take-off or landing roll. The term is often expanded to cover other wildlife
strikes - with bats or ground animals. Bird Strike is common and can be a significant
threat to aircraft safety. For smaller aircraft, significant damage may be caused to the
aircraft structure and all aircraft, especially jet-engine ones, are vulnerable to the loss of
thrust which can follow the ingestion of birds into engine air intakes. This has resulted in
several fatal accidents. Bird strikes may occur during any phase of flight, but are most
likely during the take-off, initial climb, approach and landing phases due to the greater
numbers of birds in flight at lower levels. To have a closer look the following document
visually depicts the data collected on Bird Strikes by FAA between 2000-2011.

# Bird Strike in US

```python
# Load US states geometry data from Census Bureau
url = "https://www2.census.gov/geo/tiger/GENZ2020/shp/cb_2020_us_state_20m.zip"
states = gpd.read_file(url)

# Remove Alaska, Hawaii, and territories for continental US visualization
states = states[~states['STUSPS'].isin(['AK', 'HI', 'PR', 'GU', 'VI', 'MP', 'AS'])]

# Convert state names to lowercase for matching
states['name'] = states['NAME'].str.lower()


# Assuming df is your bird strikes dataset
# Create state-wise count
df1 = df.groupby('origin_state').size().reset_index(name='n')
df1['origin_state'] = df1['origin_state'].str.lower()
df1 = df1.rename(columns={'origin_state': 'name'})

# Merge with geodataframe
bird_map = states.merge(df1, on='name', how='left')
```

```python
# Create the map
fig, ax = plt.subplots(1, 1, figsize=(10, 6), facecolor='#E8E8E8')
ax.set_facecolor('#E8E8E8')

# Plot the choropleth
bird_map.plot(column='n',
              ax=ax,
              legend=True,
              legend_kwds={'label': 'Number of Bird Strikes',
                          'orientation': 'horizontal'},
              missing_kwds={'color': 'lightgrey'},
              cmap='YlOrRd',
              edgecolor='white')

# Customize the map
ax.set_title('Bird Strikes in US\n2000-2011', pad=20)
ax.axis('off')

plt.tight_layout()
plt.show()
```

ðŸ“Œ **California, Texas and Florida has the highest number of bird strike incidents.**

## When do most bird strikes occur

```python
# Filter out rows with missing 'year', group by 'year' and count occurrences of 'conditions_sky'
sky_condition_counts = (
    df
    .loc[df['year'].notna()]  # Filter out rows with NA in 'year'
    .groupby(['year', 'conditions_sky'])
    .size()
    .reset_index(name='n')  # Reset index to get a DataFrame
)

# Create the facet bar plot with increased size and 3 plots per row
g = sns.catplot(
    data=sky_condition_counts,
    x='n',
    y='conditions_sky',
    col='year',
    kind='bar',
    height=5,           # Height of each individual plot
    aspect=1.5,         # Aspect ratio of each plot
    col_wrap=3,         # Wrap the plots into 3 per row
    palette='muted'     # Change palette as needed
)

# Add value labels on bars, shift them to the left or right
for ax in g.axes.flat:
    for p in ax.patches:
        # Adjust the label position (shift by -0.2 for left, +0.2 for right)
        x_offset = -90  # Change this value to shift left or right
        y_position = p.get_y() + p.get_height() / 2  # Center vertically
        
        ax.annotate(f'{p.get_width()}', 
                    (p.get_width() + x_offset, y_position),  # Shift label
                    ha='center', va='center', fontsize=12, color='white', fontweight='bold')

# Customize plot title
g.fig.suptitle('Bird Strike Incidents by Condition of Sky', fontsize=24)

# Set background color
for ax in g.axes.flat:
    ax.set_facecolor('#E8E8E8')  # Set background color for each subplot

g.fig.set_facecolor('#E8E8E8')  # Set figure background color

# Adjust layout
plt.subplots_adjust(top=0.9)  # Adjust the title position
plt.show()
```

## Altitude of Airplane at the time of bird strike

```python
# Filter out rows with NA in 'altitude_bin', count occurrences, and calculate percentages
altitude_counts = (
    df
    .loc[df['altitude_bin'].notna()]  # Filter out rows with NA in 'altitude_bin'
    .groupby('altitude_bin')
    .size()
    .reset_index(name='n')  # Reset index to get a DataFrame
)

# Calculate total and percentage
altitude_counts['total'] = altitude_counts['n'].sum()
altitude_counts['per'] = altitude_counts['n'] / altitude_counts['total']

# Create the bar plot for altitude of airplane
plt.figure(figsize=(10, 6))  # Set the figure size
sns.barplot(data=altitude_counts, x='altitude_bin', y='per', palette='Set2')

# Add percentage labels on the bars, with manual shifting
for index, row in altitude_counts.iterrows():
    # Adjust the label position
    x_position = index  # Base position on the index
    y_position = row['per']  # Y position at the top of the bar
    
    plt.text(x_position , y_position, 
             f'{round(row["per"] * 100, 2)}%', 
             color='black', ha='center', va='bottom', fontsize=10)

# Customize plot
plt.ylim(0, altitude_counts['per'].max() * 1.1)  # Extend y-axis to fit labels
plt.ylabel('Percentage of Incidents')
plt.xlabel('Altitude Bin')
plt.title('Percent of Incidents by Altitude of Airplane at the Time of Strike')

# Set background color for the plot
ax = plt.gca()
ax.set_facecolor('#E8E8E8')  # Set the axes background color
plt.gcf().set_facecolor('#E8E8E8')  # Set the figure background color

# Show the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

ðŸ“Œ **80.84% of bird strike incidents have happened when the altitude of airplane was <1000 ft and 19.16% have happend when altitude was >1000 ft.**


```python
# Filter out rows with NA in 'altitude_bin', group by year and count altitude_bin occurrences
altitude_yearly = (
    df
    .loc[df['altitude_bin'].notna()]  # Filter out rows with NA in 'altitude_bin'
    .groupby(['year', 'altitude_bin'])
    .size()
    .reset_index(name='n')  # Reset index to get a DataFrame
)

# Calculate total for each year and percentage
altitude_yearly['total'] = altitude_yearly.groupby('year')['n'].transform('sum')
altitude_yearly['per'] = altitude_yearly['n'] / altitude_yearly['total']

# Create the facet bar plot for altitude bins per year
g = sns.FacetGrid(altitude_yearly, col='year', col_wrap=3, height=4, sharey=False)

# Map the bar plot onto the grid
g.map_dataframe(sns.barplot, x='altitude_bin', y='per', palette='Set2')

# Add percentage labels on the bars, shifted downwards
for ax in g.axes.flat:
    for p in ax.patches:
        # Shift label down by subtracting a value (e.g., 0.02)
        ax.annotate(f'{round(p.get_height() * 100, 2)}%', 
                    (p.get_x() + p.get_width() / 2., p.get_height() + 0.02),  # Adjust this value as needed
                    ha='center', va='bottom', color='black', fontsize=10)

# Customize y-axis to show percentages
g.set(ylim=(0, altitude_yearly['per'].max() * 1.1))
g.set_axis_labels("Altitude Bin", "Percentage of Incidents")
g.set_titles(col_template="{col_name}")  # Title for each facet
g.fig.suptitle("Percent of Incidents by Altitude of Airplane at the Time of Strike per Year", fontsize=16)

# Set background color for each facet and the entire figure
for ax in g.axes.flat:
    ax.set_facecolor('#E8E8E8')  # Set the axes background color
g.fig.set_facecolor('#E8E8E8')  # Set the figure background color

# Adjust layout
plt.subplots_adjust(top=0.9)  # Adjust top for main title
plt.show()
```

## Phase of flight at the time of strike

```python
# Filter out rows with NA in 'when_phase_of_flight' and count occurrences
flight_phase_counts = (
    df
    .loc[df['when_phase_of_flight'].notna()]
    .groupby('when_phase_of_flight')
    .size()
    .reset_index(name='n')
)

# Calculate percentages directly
total = flight_phase_counts['n'].sum()
flight_phase_counts['per'] = flight_phase_counts['n'] / total

# Create ordered list based on your actual distribution
correct_order = ['Approach', 'Landing Roll', 'Take-off run', 'Climb', 'Descent', 'Taxi', 'Parked']

# Reorder the dataframe
flight_phase_counts = flight_phase_counts.set_index('when_phase_of_flight').loc[correct_order].reset_index()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=flight_phase_counts, 
            x='per', 
            y='when_phase_of_flight', 
            palette='Set2')

# Add percentage labels
for i, row in flight_phase_counts.iterrows():
    percentage = row['per'] * 100
    if percentage > 2:  # Threshold for inside/outside label placement
        plt.text(row['per']/2,  
                i,  
                f'{percentage:.1f}%',
                ha='center',
                va='center',
                color='black',
                fontweight='bold',
                fontsize=10)
    else:
        plt.text(row['per'] + 0.01,
                i,
                f'{percentage:.1f}%',
                ha='left',
                va='center',
                color='black',
                fontweight='bold',
                fontsize=10)

# Customize the plot
plt.xlim(0, flight_phase_counts['per'].max() * 1.15)
plt.ylabel('Phase of Flight')
plt.xlabel('Percentage of Incidents')
plt.title('Percent of Incidents According to Phase of Flight at the Time of Strike')

# Set background color and grid
plt.gca().set_facecolor('#E8E8E8')
plt.gcf().set_facecolor('#E8E8E8')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Format x-axis as percentages
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0f}%'.format(x*100)))

plt.tight_layout()
plt.show()
```

```python
# Filter data for warned pilots
warned_df = df[df['pilot_warned_of_birds_or_wildlife'] == 'Y']

# Calculate percentages by year and damage effect
yearly_damage = (warned_df.groupby(['year', 'effect_indicated_damage'])
                .size()
                .reset_index(name='n'))

# Calculate percentages within each year
yearly_damage['total'] = yearly_damage.groupby('year')['n'].transform('sum')
yearly_damage['per'] = yearly_damage['n'] / yearly_damage['total']

# Create the plot with overall background color
plt.figure(figsize=(15, 8), facecolor='#E8E8E8')  # Set the overall figure background color

# Create subplots for each year
g = sns.FacetGrid(yearly_damage, col='year', col_wrap=3)

# Iterate through each subplot to plot bars
for ax, (year, data) in zip(g.axes.flat, yearly_damage.groupby('year')):
    # Sort data by percentage in descending order
    sorted_data = data.sort_values('per', ascending=False)
    
    # Create the bar plot
    sns.barplot(data=sorted_data, 
                x='effect_indicated_damage', 
                y='per',
                palette='Set2', 
                ax=ax)
    
    # Add percentage labels on bars
    for i, v in enumerate(sorted_data['per']):
        ax.text(i, v + 0.02, f'{v * 100:.2f}%',  # Adjust vertical position of text
                ha='center', 
                va='bottom',
                color='black',  # Change to black for better visibility
                size=10)  # Remove fontweight to make it normal
    

    # Rotate x-axis labels for better readability
    ax.set_xticklabels(sorted_data['effect_indicated_damage'], rotation=45, ha='right')
    ax.set_xlabel('Effect Damage', fontsize=12)
    ax.set_ylabel('Percentage', fontsize=12)
    ax.set_facecolor('#E8E8E8')  # Set the axes background color

# Customize the plot title
plt.suptitle('Percent of Incidents Where Pilot Was Warned About Birds', y=1.02, size=16)

# Format y-axis as percentages
for ax in g.axes.flat:
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_yticks([0, 0.25, 0.5, 0.75, 1])  # Set specific ticks
    ax.set_yticklabels(['0%', '25%', '50%', '75%', '100%'], fontsize=10)  # Custom y-tick labels

# Display year labels as bold integers
for ax in g.axes.flat:
    year_label = ax.get_title().split('=')[1].strip()  # Extract year from title
    ax.set_title(year_label, fontweight='bold', fontsize=14)

# Adjust layout
plt.tight_layout()
plt.show()
```

## Wildlife Number Struck

```python
# Filter the DataFrame for non-null wildlife_number_struck
filtered_df = df[df['wildlife_number_struck'].notna()]

# Group by wildlife_number_struck and effect_indicated_damage and count occurrences
wildlife_counts = (
    filtered_df
    .groupby(['wildlife_number_struck', 'effect_indicated_damage'])
    .size()
    .reset_index(name='n')
)

# Calculate percentages within each wildlife_number_struck group
wildlife_counts['total'] = wildlife_counts.groupby('wildlife_number_struck')['n'].transform('sum')
wildlife_counts['per'] = wildlife_counts['n'] / wildlife_counts['total']

# Create the figure with the background color
plt.figure(figsize=(12, 8))

# Create the FacetGrid
g = sns.FacetGrid(wildlife_counts, 
                  col='wildlife_number_struck', 
                  height=4, 
                  col_wrap=2,
                  despine=False)

# Create a bar plot for each facet
g.map_dataframe(sns.barplot, 
                x='effect_indicated_damage', 
                y='per', 
                palette='Set2')

# Get unique wildlife number struck values
unique_groups = wildlife_counts['wildlife_number_struck'].unique()

# Customize each subplot
for idx, ax in enumerate(g.axes.flat):
    # Set background color for the subplot
    ax.set_facecolor('#E8E8E8')
    
    if idx < len(unique_groups):  # Check if we still have data to plot
        current_group = unique_groups[idx]
        current_data = wildlife_counts[wildlife_counts['wildlife_number_struck'] == current_group]
        
        # Add percentage labels on the bars
        for i, row in current_data.iterrows():
            bar_idx = list(current_data['effect_indicated_damage']).index(row['effect_indicated_damage'])
            ax.text(bar_idx, row['per'],
                    f'{row["per"]*100:.1f}%',
                    ha='center',
                    va='bottom')
    
    # Format x-axis labels
    ax.set_xlabel('Effect Indicated Damage')
    
    # Format y-axis as percentages
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    
    # Add gridlines
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Set the figure background color
g.fig.patch.set_facecolor('#E8E8E8')

# Customize the overall plot
g.set_axis_labels('Effect Indicated Damage', 'Percentage of Incidents')
g.set_titles(col_template='Wildlife Number Struck = {col_name}')

# Add main title
g.fig.suptitle('Percent of Incidents According to Wildlife Number Struck', 
               fontsize=16, 
               y=1.02)

# Adjust layout
plt.tight_layout()
plt.show()
```
